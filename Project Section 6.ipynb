```python
# SECTION 6.2 - Creating and Transforming DataFrames
```

```python
dict_vehicles = {"Type": ["Car","SUV","Truck","Scooter","Bike"],
           "Cost": ["500","600","800","250","75"]}
```

```python
df_vehicles = pd.DataFrame(dict_vehicles, columns = ["Type","Cost"])
```

```python
df_vehicles
```

```python
df_flip = df_vehicles.T
```

```python
df_flip
```

```python
df_flip = df_flip.rename(columns = {
    df_flip.columns[0]: "Option 1",
    df_flip.columns[1]: "Option 2",
    df_flip.columns[2]: "Option 3",
    df_flip.columns[3]: "Option 4",
    df_flip.columns[4]: "Option 5",
})
```

```python
df_flip
```

```python
df_flipback = df_flip.T
```

```python
df_flipback
```

```python
df_flipback = df_flipback.reset_index() # make the index column a column of values, and add another index column to the left.
```

```python
df_flipback
```

```python
df_flipback = df_flipback.rename(columns = {df_flipback.columns[0]: "Option"})
```

```python
df_flipback
```

```python
# add a sale price column 25% off all options
df_flipback["Sale price"] = df_flipback["Cost"].astype(float) * 0.75
```
