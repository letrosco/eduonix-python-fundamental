```python
mpg_by_origin_year = df_mpg.groupby(['origin','model_year'])['horsepower'].mean()
```

```python
df_mpg_by_origin_year = mpg_by_origin_year.reset_index()
```

```python
df_mpg_by_origin_year.origin
```

```python
all = 'ALL'
def unique_sorted_values_plus_ALL(array):
    '''
    This is used for creating an options list of unique options from a list.
    Pass in a column.
    '''
    unique = array.unique().tolist() # get a list of the unique values from the input array, and convert it into a list
    unique.sort() # sort the values
    unique.insert(0,all) # add the option ALL to the top of the list.
    return unique # return the list

# build the dropdown
dropdown_country = widgets.Dropdown(options = unique_sorted_values_plus_ALL(df_mpg_by_origin_year.origin))

# handler for the dropdown
def dropdown_country_handler(change):
    if (change.new == all):
        display(df_mpg_by_origin_year)
    else:
        display(df_mpg_by_origin_year[df_mpg_by_origin_year.origin == change.new]) # display where the column value = the dropdown value
        
# look at the output of the widget
dropdown_country.observe(dropdown_country_handler, names = 'value')
```
